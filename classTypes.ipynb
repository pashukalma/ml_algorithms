Let $y = f(x; \theta)$ be the ML algorithm, parameterized by the set of learnable
parameters $\theta$ of the ML algorithm. We define the classification loss
as the number of misclassified examples:

**Loss function**

$$ L(\theta) = \frac{1}{n} \sum_{i=1}^{n} \mathbb{1}(y_i \neq f(x_i, \theta)) $$




For continuous labels or response variables, a common loss function is the MSE. By taking the square, we eliminate the negative sign and penalize large deviations from ground truth.

$$ L(\theta) = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - f(x_i, \theta) \right)^2 $$

In **Supervised learning**, generalize on the unseen example by achieving the high accuracy, low loss function, not just on the training data but also in the new, unseen, test data.

Project high-dimensional data into lower dimensional to project patterns. Data points are represented in cluster with a data compression algorithm.

Any point not within proximity of the cluster is considered anomaly detection. Clusters of data points projected in low dimensional space would be with **unsupervised learning algorithms**. They also includes Dimensionality Reduction (e.g., PCA) and Association Rules, not just clustering.

**A probabilistic model** is described by a set of parameters theta, models as r.v.s that control the behavior of the model and its input associated data x.

**The Bayesian Inference** finds the posterior distribution p(theta|x) and capturea particular aspect of reality. The prior $p(\theta)$ is out initial belief informative or not. Our inference depends on the prior we choose -- the value of the prior parameter and the functional form of the prior. This is a chain of updates in which the prior becomes the posterior. **The Bayes engine** (Sequential Bayesian Estimation) showing the transformation of prior to a posterior as more data is observed.

We model the relationships between rv.s as a graph giving rise to **probabilistic graph models**. Each rv represents a node, and each dependency an edge. The topology of the graph changes however the goals of Bayesian inference remains the same - find the posteriors over model parameters given the observations.

The $p(x) = \int p(x \mid \theta) p(\theta) \, \mathrm{d}\theta$ is often the bottleneck that necessitates MCMC or VI.

```python
""" Sequential Bayesian Parameter Estimation - the prior becomes the posterior through a chain of updates. """
```



$$ p(\theta | x) = \frac{p(x | \theta) p(\theta)}{p(x)} = \frac{p(x | \theta) p(\theta)}{\int p(x | \theta) p(\theta) d\theta} \propto p(x | \theta) p(\theta) $$

**MCMC algorithm**

$$ \{\theta^{(t)}\}_{t=1}^{T} = \theta^{(1)}, \theta^{(2)}, \dots, \theta^{(T)} $$

$$ p(\theta | x) = \int p(\theta' | x) K(\theta | \theta') d\theta' $$

$$ E[g(\theta) | x] \approx \frac{1}{T - B} \sum_{t=B+1}^{T} g(\theta^{(t)}) $$

KL Divergence

$$ D_{KL}(p \parallel q) = \mathbb{E}_{x \sim p} \left[ \log \frac{p(x)}{q(x)} \right] $$
$$ \min_{q} D_{KL}(q(\theta) \parallel p(\theta | x)) $$

**MCMC vs. Variational Inference (VI)**
in practice, we usually maximize the ELBO (Evidence Lower Bound) because the actual KL divergence involves the intractable $p(x)$.

$$\text{ELBO}(q) = \mathbb{E}_{q(\theta)}[\log p(x, \theta) - \log q(\theta)]$$

###Problem solving paradigms:


**Complete search** traversing the entire search space for the solution. ML is an exact inference by complete enumeration **Greedy approach** taking a locally optimum choice each step and eventually reaching the global minimum solution **Divide and conquer** dividing the problem into smaller, independent problems  combining their solution. **Dynamic programming** dividing the problem into smaller, independent problems, compute the solution for each, and store it in the DP table.
